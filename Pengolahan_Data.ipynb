```python
import numpy as np

# Membaca file CSV
data = np.genfromtxt('Data1.csv', delimiter=';', skip_header=1, usecols=(0, 1), dtype=str)

# Mengganti koma (',') dengan titik ('.')
data = np.char.replace(data, ',', '.')

# Mengonversi tipe data menjadi float
data = data.astype(float)
```

```python
print(data)
```

```python
print(data.T[0])
```

```python
print(data.T[1])
```

```python
plt.figure(figsize=(10,6))
plt.plot(data.T[0], data.T[1])
plt.grid()
plt.xlabel('Waktu (s)')
plt.ylabel('Perubahan Percepatan narik ikan(m/s^2)')
plt.title('Grafik Perubahan Percepatan Memancing ')
```
